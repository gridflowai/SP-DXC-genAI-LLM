```python
import pandas as pd
import numpy  as np

import matplotlib.pyplot as plt

import seaborn as sns
sns.set(style="white", color_codes=True)

from sklearn.datasets  import load_breast_cancer

np.set_printoptions(edgeitems=8, precision=4, suppress=True, linewidth=140)
```

#### load the data (breastcancer data)

```python
breast_cancer_data = load_breast_cancer()
```

```python
df_X = pd.DataFrame(data=breast_cancer_data.data, columns = breast_cancer_data.feature_names)
df_y = pd.DataFrame(data=breast_cancer_data.target, columns = ['target'])
```

#### Prepare the data splits

```python
df_X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=.20)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#### build the ML model (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier     # import the lib
```

```python
knn = KNeighborsClassifier(n_neighbors=5)              # instantiate the model
```

```python
knn.fit(X_train, y_train)                              # Fit the data to the model (train the model)
```

```python
y_pred = knn.predict(X_test)                           # predict (test samples)
```

#### Evaluate the model (metrics)

```python
y_test.target.values
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```

**Confusion Matrix**

```python
metrics.confusion_matrix(y_test, y_pred)
```

**classification report**

```python
print(metrics.classification_report(y_test, y_pred))
```
