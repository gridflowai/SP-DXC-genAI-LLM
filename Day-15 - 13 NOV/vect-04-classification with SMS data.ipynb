```python
import numpy as np
import pandas as pd
```

```python
location = r'D:\AI-DATASETS\01-MISC\sms.tsv'
```

```python
df = pd.read_csv(location, sep='\t', header=None, names = ['target', 'sms_text'])
```

```python
df.sample(10)
```

```python
df.shape
```

```python
X = df.sms_text
y = df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20)
```

#### Vectorize the train data

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer()
```

```python
vect.fit(X_train)
```

```python
len(vect.get_feature_names_out())
```

#### transform the training set

```python
X_train_vect = vect.transform(X_train)
```

#### build a ML model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train_vect, y_train)
```

#### test the ML model with X_test

```python
X_test_vect = vect.transform(X_test)
```

```python
y_pred = knn.predict(X_test_vect)
```

#### compare the predictions vs the actuals

```python
from sklearn import metrics
```

```python
y_test, y_pred
```

```python
# Accuracy score
metrics.accuracy_score(y_test, y_pred)
```

```python
# confusion matrix
metrics.confusion_matrix(y_test, y_pred)
```

```python
# Classification report
print(metrics.classification_report(y_test, y_pred))
```

#### Change the ML model (use naive bayes)

```python
from sklearn.naive_bayes import  MultinomialNB
```

```python
mnb = MultinomialNB()

mnb.fit(X_train_vect, y_train)

y_pred = mnb.predict(X_test_vect)

print(metrics.accuracy_score(y_test, y_pred))

print(metrics.confusion_matrix(y_test, y_pred))
```
