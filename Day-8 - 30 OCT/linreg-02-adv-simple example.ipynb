```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

#### Load the data

- advertising data

```python
location = "https://github.com/gridflowai/gridflowAI-datasets-icons/raw/master/AI-DATASETS/01-MISC/Advertising.csv"
```

```python
df = pd.read_csv(location)
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.sample(5)
```

```python
feature_cols = ['TV',	'radio',	'newspaper']
```

```python
X = df[feature_cols]
y = df['sales']
```

I - (import the ML library)

```python
from sklearn.linear_model import LinearRegression
```

I - (Instantiate the model)

```python
linreg = LinearRegression()
```

F - (fit the data  to the model, train the model)

```python
linreg.fit(X, y)
```

```python
linreg.coef_                  # m1, m2, m3
```

```python
linreg.intercept_             # C 
```

P - Predict (for new data)

```python
new_data = [[10, 5, 6]]
```

```python
linreg.predict(new_data)
```
