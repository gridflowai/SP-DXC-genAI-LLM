```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define words for each category
animals = ["cat", "dog", "elephant", "tiger", "lion"]
cities  = ["paris", "london", "tokyo", "newyork", "sydney"]
books   = ["harrypotter", "lordoftherings", "gatsby", "mockingbird", "dune"]
```

```python
# Generate random 2D embeddings for each category
np.random.seed(42)  # For reproducibility
animal_embeddings = np.random.rand(len(animals), 2)  + np.array([0, 0])    # Shift embeddings for animals
city_embeddings   = np.random.rand(len(cities),  2)  + np.array([3, 3])    # Shift embeddings for cities
book_embeddings   = np.random.rand(len(books),   2)  + np.array([6, 6])    # Shift embeddings for books
```

```python
# Plotting the embeddings
plt.figure(figsize=(8, 8))
# Plot animals
for i, word in enumerate(animals):
    plt.scatter(animal_embeddings[i, 0], animal_embeddings[i, 1], color='blue', label='Animals' if i == 0 else "")
    plt.text(animal_embeddings[i, 0] + 0.1, animal_embeddings[i, 1], word, fontsize=12)

# Plot cities
for i, word in enumerate(cities):
    plt.scatter(city_embeddings[i, 0], city_embeddings[i, 1], color='green', label='Cities' if i == 0 else "")
    plt.text(city_embeddings[i, 0] + 0.1, city_embeddings[i, 1], word, fontsize=12)

# Plot books
for i, word in enumerate(books):
    plt.scatter(book_embeddings[i, 0], book_embeddings[i, 1], color='red', label='Books' if i == 0 else "")
    plt.text(book_embeddings[i, 0] + 0.1, book_embeddings[i, 1], word, fontsize=12)

# Adding labels and title
plt.xlabel('Embedding Dimension 1')
plt.ylabel('Embedding Dimension 2')
plt.title('2D Embeddings for Animals, Cities, and Books')
plt.legend()
plt.grid(True)
plt.show()
```
