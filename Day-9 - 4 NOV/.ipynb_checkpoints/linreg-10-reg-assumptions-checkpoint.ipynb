#### 3. checking for residuals (y_test - y_pred) - random

```python
location = "https://github.com/gridflowai/gridflowAI-datasets-icons/raw/master/AI-DATASETS/01-MISC/Advertising.csv"
```

```python
# load the training data from glass data set
df_training = pd.read_csv(location)
```

```python
df_training.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# create a Python list of feature names
feature_cols = ['TV', 'radio', 'newspaper']

# use the list to select a subset of the original DataFrame
X = df_training[feature_cols]

# select a Series from the DataFrame
y = df_training['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()

# train the model (fit the data to the model)

linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
(y_pred - y_test).mean()          # close to 0 would mean that residuals are random
```

```python
residual = y_pred - y_test
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3 , figsize=(16, 6))
ax1.scatter(y_pred, residual)
ax1.set_xlabel('predictions')
ax1.set_ylabel('residual')

ax2.scatter(y_test, residual)
ax2.set_xlabel('actual')
ax2.set_ylabel('residual')

ax3.scatter(y_test, y_pred)
ax3.set_xlabel('actual')
ax3.set_ylabel('predictions');
```
